```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import geopandas as gpd
```

```python
slc = pd.read_csv('./SLC_airbnb_data.csv')
```

```python
# select salt lake city area only
slc_only = slc.query('-111.99 < long < -111.79 and 40.7 < lat < 40.825')
```

```python
# Read in the services district shapefile and look at the first few rows
slc_map = gpd.read_file('./Atlas_Plat_Index/Atlas_Plat_Index.shp')
print(slc_map.head())
```

```python
# salt lake city
slc_map.to_crs(epsg=4326).plot(figsize=(10, 10));
```

```python
slc_map.to_crs(epsg=4326).plot(figsize=(10, 10));

# Add the airbnb locations, by price groups
sns.scatterplot(slc_only.long, slc_only.lat,hue=slc_only.price_category, palette='husl', s=20)
```

```python
slc_map.to_crs(epsg=4326).plot(figsize=(10, 10));

# Add the airbnb locations, by number of bed
sns.scatterplot(slc_only.long, slc_only.lat,hue=slc_only.Bed, s=15, palette='Set2')
plt.show()
```

```python
slc_map.to_crs(epsg=4326).plot(figsize=(10, 10));
# Add the airbnb locations, by house types
sns.scatterplot(slc_only.long, slc_only.lat,hue=slc_only.House_type, s=10)
```
