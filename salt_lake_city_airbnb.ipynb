```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
slc = pd.read_csv('./SLC_airbnb_data.csv')
```

```python
# checking type of every column in the dataset
slc.dtypes
```

```python
#looking to find out first what columns have null values
#using 'sum' function will show us how many nulls are found in each column in dataset
slc.isnull().sum()
```

```python
# examining the unique values of 'House_type'
slc.House_type.unique()
```

```python
# examining the unique values of 'category'
slc.category.unique()
```

#### Correlation between different variables


```python
corr = slc.corr(method='kendall')
plt.figure(figsize=(15,8))
sns.heatmap(corr, annot=True)
slc.columns
```

#### Data Visualization

```python
# setting figure size for future visualizations
sns.set(rc={'figure.figsize':(10,8)})
sns.set_style('white')
```

```python
viz_price=slc.plot(kind='scatter', x='long', y='lat',c='Price_usd',
                  cmap=plt.get_cmap('jet'), colorbar=True, alpha=0.4, figsize=(10,8))
viz_price.legend()
```

```python
sns.countplot(slc['Location_city'], palette="plasma")
fig = plt.gcf()
fig.set_size_inches(10,10)
plt.xticks(rotation=75)
plt.title('Location_city')
```

```python
sns.countplot(slc['House_type'], palette="plasma", order=slc['House_type'].value_counts().index)
fig = plt.gcf()
fig.set_size_inches(10,10)
plt.xticks(rotation=75)
plt.title('House_type')
```

```python
sns.countplot(slc['Room Type'], palette="plasma", order=slc['Room Type'].value_counts().index)
fig = plt.gcf()
fig.set_size_inches(10,10)
plt.xticks(rotation=75)
plt.title('Room Type')
plt.show()
```

```python
sns.countplot(slc['price_category'], palette="plasma", order=slc['price_category'].value_counts().index)
fig = plt.gcf()
fig.set_size_inches(10,10)
plt.xticks(rotation=75)
plt.title('Price Category')
plt.show()
```

```python
sns.countplot(slc['Guest_Maximum'], palette="plasma")
fig = plt.gcf()
fig.set_size_inches(10,10)
plt.xticks(rotation=75)
plt.title('Guest Maximum')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(slc.long, slc.lat,hue=slc.House_type)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(slc.long, slc.lat,hue=slc.Bed)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(slc.long, slc.lat,hue=slc.price_category, palette='husl')
plt.show()
```

```python
under_500 = slc[slc['Price_usd'] < 500]
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(under_500['ratings'], under_500['Price_usd'],hue=slc.Bed)
plt.show()
```
